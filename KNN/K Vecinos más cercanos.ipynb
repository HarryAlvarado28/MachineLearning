# K Vecinos más cercanos

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

## Obtención de Datos

```python
dataframe = pd.read_csv('vehiculos.csv')
dataframe.head()
```

```python
dataframe['vehicle_class'].unique()
```

## Limpieza de Datos

```python
y = dataframe['vehicle_class'] # Variable Objetivo
x = dataframe.drop('vehicle_class',axis=1) # Variables Caracteristicas
```

## Datos de Entrenamiento - *Datos de Pruebas*

```python
from sklearn.model_selection import train_test_split
```

```python
x_train, x_test, y_train, y_test = train_test_split(x,y,test_size=0.3,random_state=45)
x_train
```

## Modelo de Entrenamiento

```python
from sklearn.neighbors import KNeighborsClassifier
knn = KNeighborsClassifier(n_neighbors=1)
```

```python
knn.fit(x_train,y_train)
```

## Modelo de Pruebas

```python
predicciones = knn.predict(x_test)
predicciones
```

```python
from sklearn.metrics import classification_report, confusion_matrix
print(confusion_matrix(y_test, predicciones))
```

```python
print(classification_report(y_test,predicciones))
```

Despues de cierta reviciones se convierte en el **Modelo Definitivo**

```python
tasa_error = []
for i in range(1,20):
    knn = KNeighborsClassifier(n_neighbors=i)
    knn.fit(x_train, y_train)
    prediccion_i = knn.predict(x_test)
    tasa_error.append(np.mean(prediccion_i != y_test))
    
tasa_error
```

```python
knn = KNeighborsClassifier(n_neighbors=3)
knn.fit(x_train, y_train)
predicciones = knn.predict(x_test)
print(classification_report(y_test,predicciones))
```
