# Gr√°ficos con Pandas

```python
import numpy as np
import pandas as pd
```

```python
dataframe = pd.DataFrame(np.random.randint(200,size=(50,4)),columns=['a','b','c','d'])
dataframe
```

## Gr√°ficos de Hist

```python
dataframe['a'].hist()
```

```python
dataframe['b'].hist()
```

```python
dataframe['b'].hist(bins=30)
```

```python
dataframe['a'].plot.hist()
```

## Gr√°ficos de √Årea

```python
dataframe.plot.area(alpha=.6)
```

## Gr√°ficos de Barra

```python
dataframe.plot.bar(stacked=True)
```

## Gr√°ficos de Scatter

```python
dataframe.plot.scatter(x='a',y='b',c='c',cmap='coolwarm')
```

## Gr√°ficos de Caja (box)

```python
dataframe.plot.box()
```

## Gr√°ficos de Hexbin

```python
dataframe.plot.hexbin(x='a',y='b',gridsize=12)
```

## Gr√°ficos de KDE

```python
dataframe.plot.kde()
```

## Gr√°ficos de Density

```python
dataframe.plot.density()
```
