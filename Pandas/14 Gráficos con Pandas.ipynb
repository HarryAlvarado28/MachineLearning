# Gráficos con Pandas

```python
import numpy as np
import pandas as pd
```

```python
dataframe = pd.DataFrame(np.random.randint(200,size=(50,4)),columns=['a','b','c','d'])
dataframe
```

## Gráficos de Hist

```python
dataframe['a'].hist()
```

```python
dataframe['b'].hist()
```

```python
dataframe['b'].hist(bins=30)
```

```python
dataframe['a'].plot.hist()
```

## Gráficos de Área

```python
dataframe.plot.area(alpha=.6)
```

## Gráficos de Barra

```python
dataframe.plot.bar(stacked=True)
```

## Gráficos de Scatter

```python
dataframe.plot.scatter(x='a',y='b',c='c',cmap='coolwarm')
```

## Gráficos de Caja (box)

```python
dataframe.plot.box()
```

## Gráficos de Hexbin

```python
dataframe.plot.hexbin(x='a',y='b',gridsize=12)
```

## Gráficos de KDE

```python
dataframe.plot.kde()
```

## Gráficos de Density

```python
dataframe.plot.density()
```
