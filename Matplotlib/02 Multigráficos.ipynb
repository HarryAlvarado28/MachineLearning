# Multigr치ficos 

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.linspace(2,9,10)
x
```

```python
y = x ** 3
y
```

```python
figura, graficos = plt.subplots(nrows=2, ncols=2)

graficos[0][0].plot(x,y,'blue')
graficos[0][1].plot(x,y,'red')
graficos[1][0].plot(x,y,'y')
graficos[1][1].plot(x,y,'g')

graficos[0][0].set_title('Azul')
graficos[0][1].set_title('Rojo')
graficos[1][0].set_title('Amarillo')
graficos[1][1].set_title('Verde')
```

# Tama침o del Gr치fico

```python
figure = plt.figure(figsize=(12,2))
grafico = figure.add_axes([0.1,0.1,0.9,0.9])
grafico.plot(x,y,'r')
```

# Dos gr치ficos en la misma figura

```python
figura = plt.figure()
graficos = figura.add_axes([0.1,0.1,0.9,0.9])
graficos.plot(x,x**2, label='A 2췈')
graficos.plot(x,x**3, label='A 3췈')
graficos.plot(x,x**4, label='A 4췈')
graficos.legend(loc=0)
```

# Color del gr치ficos

```python
figura = plt.figure()
grafico = figura.add_axes([0,0,1,1])
grafico.plot(x,y,'tomato')
```

# Linea del gr치ficos (grosor, transparencia y tipo)

```python
figura = plt.figure()
grafico = figura.add_axes([0,0,1,1])
grafico.plot(x,y,color='red',linewidth=10,alpha=0.7, linestyle=':')
```

# Marcadores 

```python
figura = plt.figure()
grafico = figura.add_axes([0,0,1,1])
grafico.plot(x,y,color='red',marker='*',markersize=15)
```
